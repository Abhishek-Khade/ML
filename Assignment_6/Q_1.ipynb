```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("ToyotaCorolla.csv",encoding='ISO-8859-1')
```

```python
data.info()
```

```python
df = data[["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight","Price"]]
```

```python
df
```

```python
df.info()
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['Price'],axis=1)
y = df['Price']
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y,shuffle=True,test_size=0.33,random_state=12345)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_true = y_test
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_true,y_pred)
mae
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true,y_pred)
mse
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y_true,y_pred)
rmse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_true,y_pred)
r2
```
