```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("delivery_time.csv")
```

```python
df
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
X = df.drop(['Delivery Time'],axis=1)
y = df['Delivery Time']
model = LinearRegression()
model.fit(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict(X)
```

```python
plt.scatter(df['Sorting Time'],df['Delivery Time'],color ='red')
plt.plot(df['Sorting Time'],model.predict(X),color='blue')
plt.scatter(df['Sorting Time'],model.predict(X),color='blue')

mean_arr = np.ones(y.size) *  y.mean()
plt.scatter(df['Sorting Time'],mean_arr,color='green')
plt.plot(df['Sorting Time'],mean_arr,color='green')
```
