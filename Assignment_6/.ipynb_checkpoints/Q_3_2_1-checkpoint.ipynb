```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("NewspaperData.csv")
```

```python
df
```

```python
df.drop(['Newspaper'],inplace=True,axis =1)
```

```python
df
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
model = LinearRegression()
X = df[['daily']]
y = df['sunday']
```

```python
model.fit(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict(X)
```

```python
plt.scatter(df['daily'],df['sunday'],color ='red')
plt.plot(df['daily'],model.predict(X),color='blue')
plt.scatter(df['daily'],model.predict(X),color='blue')

mean_arr = np.ones(y.size) *  y.mean()
plt.scatter(df['daily'],mean_arr,color='green')
plt.plot(df['daily'],mean_arr,color='green')
```
