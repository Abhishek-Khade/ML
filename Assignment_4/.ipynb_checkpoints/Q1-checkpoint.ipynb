```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("East_std.csv")
```

```python
df.head()
```

```python
df.drop(['cc1_miles','cc2_miles','cc3_miles','Bonus_miles','Flight_miles_12mo','Bonus_trans','Flight_trans_12','Days_since_enroll','Award?'],axis=1,inplace=True)
```

```python
df.drop(df.columns[0], axis=1,inplace =True)
```

```python
df.info()
```

```python
plt.scatter(df['Balance'],df['Qual_miles'])
```

```python
wscc = []
wsccn = []
for i in range(1,11):
    k = KMeans(n_clusters = i,random_state =12345)
    k.fit(df)
    wscc.append(k.inertia_)
    wsccn.append(k.inertia_/i)

plt.plot(range(1,11),wscc)
plt.plot(range(1,11),wsccn)
```

```python
k = KMeans(n_clusters =4)
k.fit(df)
```

```python
k.labels_
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=k.labels_)
```

```python
plt.scatter(df['Balance'][k.labels_ == 0],df['Qual_miles'][k.labels_ == 0],color = 'red', marker='x')
plt.scatter(df['Balance'][k.labels_ == 1],df['Qual_miles'][k.labels_ == 1],color = 'blue', marker='1')
plt.scatter(df['Balance'][k.labels_ == 2],df['Qual_miles'][k.labels_ == 2],color = 'green', marker='2')
plt.scatter(df['Balance'][k.labels_ == 3],df['Qual_miles'][k.labels_ == 3],color = 'yellow', marker='3')
plt.scatter(df['Balance'][k.labels_ == 4],df['Qual_miles'][k.labels_ == 4],color = 'black', marker='4')
plt.scatter(df['Balance'][k.labels_ == 5],df['Qual_miles'][k.labels_ == 5],color = 'orange', marker='o')
```
