```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("universities_std.csv")
```

```python
df.head()
df
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
df.drop(['Expenses','SFRatio','Top10'],axis=1,inplace=True)
df
```

```python
wss = []
wssn = []
for i in range(1,11):
    k = KMeans(n_clusters = i)
    k.fit(df)
    wss.append(k.inertia_)
    wssn.append(k.inertia_/i)
plt.plot(range(1,11),wss)
plt.scatter(range(1,11),wss)
plt.plot(range(1,11),wssn)
plt.scatter(range(1,11),wssn)
```

```python
k = KMeans(n_clusters=4)
k.fit(df)
```

```python
SAT = df['SAT']
Accept = df['Accept']
GradRate = df['GradRate']

# Create the 3D scatter plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(SAT, Accept, GradRate, c=k.labels_, marker='o')

# Set labels and title
ax.set_xlabel('SAT')
ax.set_ylabel('Accept')
ax.set_zlabel('GradRate')
ax.set_title('3D Scatter Plot')

plt.show()
```

```python
SAT = df['SAT']
Accept = df['Accept']
GradRate = df['GradRate']

# Create the 3D scatter plot
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(SAT[k.labels_ == 0], Accept[k.labels_ == 0], GradRate[k.labels_ == 0], marker='o',color ='red',s=100)
ax.scatter(SAT[k.labels_ == 1], Accept[k.labels_ == 1], GradRate[k.labels_ == 1], marker='x',color ='blue',s=100)
ax.scatter(SAT[k.labels_ == 2], Accept[k.labels_ == 2], GradRate[k.labels_ == 2], marker='1',color ='yellow',s=100)
ax.scatter(SAT[k.labels_ == 3], Accept[k.labels_ == 3], GradRate[k.labels_ == 3], marker='2',color ='green',s=100)

# Set labels and title
ax.set_xlabel('Murder')
ax.set_ylabel('Assault')
ax.set_zlabel('Rape')
ax.set_title('3D Scatter Plot')

plt.show()
```
